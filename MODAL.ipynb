## I. Introduction 

### I.1. Modèle physiologique vs modèle phénoménologique

><div align="justify">Au sein des modèles cardiaques, il existe une multitude d'échelles à partir delaquelle nous pouvons établir une hiérarchie. Partant de l'échelle intracellullaire, se présentent alors les échelles de la cellule, du tissu, de l'organe pour enfin aboutir à l'échelle du patient et celle de l'espèce. Une fois l'échelle choisie, il existe deux approches pour construire un modèle.</div>
><br/><div align="justify">La première consiste à utiliser des observations et des modèles convaincants à petite échelle pour modéliser l'échelle supérieure. A l'instar de la démarche adoptée en biologie fondamentale, ces <strong>données microscopiques</strong> sont utilisées pour <strong>donner du sens aux phénomènes macroscopiques</strong>. Une telle approche est avant tout centrée sur la bonne reproduction de la physiologie subjacente du problème, on parle de modèle <strong>physiologique</strong>.</div>
><br/><div align="justify">A contrario, la seconde se concentre sur les <strong>phénomènes observés à une échelle macroscopique</strong> afin d'établir un modèle conforme à ces observations. On passe alors progressivement à l'échelle microscopique en enrichissant notre modèle d'<strong>informations complémentaires</strong> permettant de combler les éventuelles insuffisances qui apparaissent avec la diminution de l'ordre de grandeur. Il s'agit dans ce cas de modèle <strong>phénoménologique</strong> qu'il est possible de comparer à la recherche clinique.</div>
><br/><div align="justify">La modélisation de l'éléctrophysiologie cardiaque utilise ces deux approches dans le but de créer des modèles relativment complets et pertinents. En effet, les modèles de propagation de potentiel d'action à l'échelle du tissu reposent sur une extrapolation de modèles établis à l'échelle cellulaire. Dans ce cas, l'approche physiologique est choise. Cependant, si on s'intéresse aux conséquences anatomiques et éléctriques induites par une filbrillation auriculaire, l'approche phénoménologique va être utilisée en s'appuyant d'observations réalisées lors d'études cliniques.</div></div>

### II. Modèle de Hodgkin-Huxley et de FitzHugh-Nagumo

#### II.1 Le modèle de Hodgkin-Huxley

><div align="justify">Hodgkin et Huxley ont publié en 1952 une série d’articles qui leur a valu le prix Nobel de Médecine et de Physiologie en 1963. Dans ces articles, ils ont notamment proposé un modèle de propagation du potentiel d’action dans l’axone géant de calmar. Ils y formalisent un bilan de courants dans la membrane ainsi que l’activité des canaux sodiques et potassiques comme un courant électrique dépendant du potentiel transmembranaire, de la conductance du canal variable de temps et des concentrations dans les milieux interne et externe de soluté entrant en jeu dans le potentiel de Nernst.
Ce modèle de propagation du potentiel d'action s'écrit : </div>

>$$
\left\{\begin{array}{l}
C_{m} \partial_{t} V_{m}=-\bar{g}_{K} n^{4}\left(V_{m}-E_{k}\right)-\bar{g}_{N a} m^{3} h\left(V_{m}-E_{N a}\right)-\bar{g}_{L}\left(V_{m}-E_{L}\right)+I_{a p p} \\
\partial_{t} n=\frac{n_{\infty}}{\tau_{n}} \\
\partial_{t} m=\frac{m_{\infty}-m}{\tau_{T} m} \\
\partial_{t} h=\frac{h_{\infty}-h}{\tau_{h}}
\end{array}\right.
$$
><br/><div align="justify">où $I_{a p p}$ est un courant appliqué par l'expérimentateur et $C_{m}$ la capacitance de la membrane, $n$ et $m$ sont des variables d'activation et $h$ est une variable d'inhibition. La dynamique des variables $n$ et $h$ est plus lente que celle de la variable $m$. Leur évolution est également qualitativement plus lente que celle de $V_{m}$.</div>

><div align="justify">Pour étudier le modèle dans le plan de phase des variables $(n, V_m)$, 
ils éliminent une variable rapide et une variable lente. L'idée est que $\forall t \ge 0$,
$m$ prend sa valeur stationnaire : $m(t, V_m) = m_{\infty}(V_m)$ et que pendant un potentiel d'action, $n+h \approx 0,8$. Ainsi, on élimine $m$ et $h$. Ce qui conduit finalement, en supposant le courant appliqué nul :</div>

>$$
\left\{\begin{array}{l}
C_{m} \partial_{t} V_{m}=-\bar{g}_{K} n^{4}\left(V_{m}-E_{k}\right)-\bar{g}_{N a} m_{\infty}\left(V_{m}\right)^{3}(0.8-n)\left(V_{m}-E_{N a}\right)-\bar{g}_{L}\left(V_{m}-E_{L}\right):=f(v, n) \\
\partial_{t} n=\frac{n_{\infty}-n}{\tau_{n}}
\end{array}\right.
$$

#### II.1 Le modéle de  FitzHugh-Nagumo

><div align="justify">FitzHugh et Nagumo proposent indépendamment une simplification de ce type.
Ils utilisent une réduction légèrement différente, mais basées sur la même idée afin d’aboutir à un problème adimensionalisé de la famille des problèmes de FitzHugh-Nagumo, c’est à dire un problème du type</div>
$$
\left\{\begin{array}{l}
\partial_{t}v = f(v) - w + I_{app} \\
\partial_{t}w = \epsilon({\beta}v - {\gamma}w)
\end{array}\right.\\
$$
avec $f(v) = v(1-v)(v-\alpha)$ pour $0\le \alpha \le 1$ et $\epsilon \ll 1.$

>On s'est intéréssé à la stabilité des solutions de ce systéme et de l'effet de seuil engendré par le seuil d'activation dans le plan de phase des variables $(v, w)$.

##### II.1.1 Résolution par la méthode Odeint

```python
"""importation des bibliothéques"""
import numpy as np
import scipy.integrate as itg
import matplotlib.pyplot as plt

"""définition des constante"""
alpha = 0.1
beta = 0.5
epsilon = 0.01
gamma = 1

"""discrétisation de la va"""
v0 = -1
vf = 1
v = np.linspace(v0,vf,100)

plt.figure(figsize = (8, 6))

# schéma des isoclines dv/dt = 0 et dw/dt = 0
plt.plot(v, (beta*v)/gamma,'g-',label ='isocline dw/dt')
plt.plot(v, v*(1-v)*(v-alpha),'g--',label = 'isocline dv/dt')

plt.xlim([-0.5, 1.3])
plt.ylim([-0.05,0.2])

#tracé de la droite du seuil d'activation 
plt.axvline(x = alpha, color='b', label ="seuil d'activation")

"""definition des variables de temps"""
t0=0
tf=500
#h=0.05
t=np.linspace(t0,tf,500)

# on definie le systeme qui renvoie un vecteur de 2 coordonnées
# correspondant aux deux dérivées des deux variables
def f_LV(y, t):
    yp=np.array([y[0]*(1-y[0])*(y[0]-alpha)-y[1], epsilon*(beta*y[0] - y[1]*gamma)])
    return yp

#on initialise deux points pour simuler deux situations differentes
#dans le 1er cas, on stimule suffisament pour créer un PA
#le 2nd cas correspond à l'orbite d'une simulation non exitatrice.
y01=np.array([alpha+0.05,0])
y02 = np.array([alpha-0.05,0])

"""on résolve le problème par odeint"""
w1 = itg.odeint(f_LV,y01, t)
w2 = itg.odeint(f_LV,y02, t)

"""on represente les orbites"""
plt.grid()
plt.plot(0,0,'o',color='black',label='point fixe')
plt.plot(w1[:,0],w1[:,1],'r',label = "orbite d'un PA")
plt.plot(w2[:,0],w2[:,1],color = 'y',label = "simulation non excitatrice")
plt.legend()
plt.xlabel('V')
plt.ylabel('W')
plt.title('Orbite de deux potentiels d’actions dans le plan des phases (v, w)')
```

><div align="justify">Une première stimulation (courbe rouge) est suffisante pour permettre à la 
variable v de dépasser le seuil d'activation $\alpha$ (droite bleue). Une fois franchit l'isocline $\partial_{t}v = 0$, elle parcourt une large boucle pour regagner le domaine stable et le point fixe. Quand à la deuxiéme simulation (courbe jaune), elle n'est pas assez forte pour dépasser le seuil d'activation. L'orbite reste alors dans la branche stable de l'isocline.</div>

```python
"""representation du potentiel d'action"""
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.plot(t, w1[:,0],label='PA en fonction du temps')
plt.ylabel('V')
ax2.plot(t, w1[:,1], label = 'variable de porte')
ax1.grid()
plt.xlabel('t')
ax2.grid()
ax1.legend(loc = 'best')
ax2.legend(loc='best')
```

#### II.1.2. Résolution par la méthode Euler implicite

><div align="justify">On vient de voir une résolution par la méthode Odeint. Cette méthode est certe puissante et trés facile à appliquer mais nous ignorons beaucoup de chose sur elle. On va alors s'intéréssé à la méthode d'Euler implicite qui demandera beaucoup plus de travail que la méthode Odeint. Il sera par la suite intéréssant de voir à quel point ces deux méthodes sont équivalentes.</div>

```python
"""Importation des bibliothéques """
import numpy as np
import matplotlib.pyplot as plt

def fct(y):
    Z = np.zeros(2)
    Z[0] = y[0]*(1-y[0])*(y[0] - alpha) - y[1]
    Z[1] = epsilon*(beta*y[0] - gamma*y[1])
    return Z

"""initialisation et choix du pas de temps"""
t0=0    
y03=np.array([alpha+0.05, 0])
y04=np.array([alpha-0.05, 0])
T=500
vect_h = np.array([0.01, 0.1, 1, 2])

"""discretisation de l'espace temps en fonction de differents pas de temps h"""


def phi(yn,h,x,f):
    return x - yn - h * f(x)

def Dphi(h,x,Df):
    dim = x.shape[0]
    return np.eye(dim) - h * Df(x)

def newton(yn,h,f,Df,tol):
    x = yn
    err = 1
    k=0
    itmax = 100
    while err > tol and k<itmax :
        x = x - np.linalg.solve(Dphi(h,x,Df), phi(yn,h,x,f))
        err = np.linalg.norm(phi(yn,h,x,f))
        k = k+1
    return x

def Dfct(y):
    return np.array([[-3*y[0]**3+2*(1+alpha)*y[0]-alpha,  0], [epsilon*beta, -epsilon*gamma]])

def sc_Euler_imp(y0,t0,T,h,f,Df):
    N=int(T/h)
    dim = y0.shape[0]
    y = np.zeros((dim,N+1))
    y[:,0]=y0
    for n in range(1,N+1):
        y[:,n]=newton(y[:,n-1], h,f,Df,1e-5)
    return y



v0 = -1
vf = 1
v = np.linspace(v0,vf,100)

fig, axs = plt.subplots(4,1, figsize=(7,30))

#representation sur quatre figure avec different pas de temps
for k in range(vect_h.size):
    N=int(T/vect_h[k])
    t=np.linspace(t0,T,N+1)
    
    w3 = sc_Euler_imp(y03, t0, T, vect_h[k], fct, Dfct)
    w4 = sc_Euler_imp(y04, t0, T, vect_h[k], fct, Dfct)
    
    axs[k].set_xlim(-1, 1)
    axs[k].set_ylim(-0.05, 0.2) 
    w3 = sc_Euler_imp(y03, t0, T, vect_h[k], fct, Dfct)
    w4 = sc_Euler_imp(y04, t0, T, vect_h[k], fct, Dfct)
    axs[k].plot(v, (beta*v)/gamma,'g--')
    axs[k].plot(v, v*(1-v)*(v-alpha),'g--')
    axs[k].axvline(x = alpha, color='b')
    axs[k].plot(w3[0,:],w3[1,:],'r',label = "orbite d'un PA")
    axs[k].plot(w4[0,:],w4[1,:],color = 'y',label = "simulation non excitatrice")
    axs[k].grid()
    plt.xlabel('V')
    plt.ylabel('W')
```

>On remarque que pour la méthode Euler implicite, on est face à des probléme de stabilité qui dépendent du pas de temps $h$. Plus h est petit plus la solution est stable.

## III. Modèle de Mitchell-Schaeffer

></br><div align="justify">Suite à ce que nous avons obtenu dans la section précédente, nous pouvons souligner que le modèle de <strong>FitzHugh-Nagumo</strong> n'est <strong>pas tout à fait bien adapté à la modélisation cardiaque</strong>. En effet, les potentiels d'action que nous obtenons correspondent à des potentiels d'action de neurone qui présentent des différences non négligeables avec les potentiels d'action cardiaque. Par exemple, on observe une <strong>hyperpolarisation</strong> caractéristique avant le retour au potentiel de repos ou encore l'<strong>abscence de phase de plateau</strong>.</div>
><br/><div align="justify">De ce fait, nous nous penchons désormais sur le modèle de Mitchell-Schaeffer qui repoe également sur le formalisme de Hodgkin-Huxley. Ce modèle est semblable au précédent, notamment avec la prise en compte d'une variable lente $h(t)$ appelée gating variable dans la littérature, et d'une variable rapide, le potentiel transmembranaire $v(t)$. Néanmoins, l'orbite d'un potentiel d'action se limite au demi-plan $v > 0$, empêchant ainsi toute hyperpolarisation à l'instar de ce que nous avons observé dans le modèle de FitzHugh-Nagumo. De plus, <strong>Mitchell-Schaeffer présente une paramétrisation de l'éléctrophysiologie subjacente</strong>, avec l'intervention de temps caractéristiques et de potentiels critiques. Enfin, ce nouveau modèle fournit également des <strong>expressions analytiques</strong> des durées de potentiels d'actions en fonctions de ces paramètres éléctrophysiologiques.</div>


### III.1 Les équations du modèle

#### III.1.1 Modélisation

><div align="justify">L'évolution du potentiel transmembranaire adimensionné <em>v</em> est régie par l'équation différentielle ordinaire :</div>
<br></br>
><div align="center">$\frac{dv}{dt} = J_{in}(v,h) + J_{out}(v) + J_{stim}(t)$</div>
><br/><div align="justify">Le courant entrant $J_{in}$  est une combinaison de tous les courants augmentant le potentiel transmembranaire, en particulier ceux générés par le déplacement de sodium et de calcium via les canaux. Son expression est donnée par :</div>
<br></br>
><br/> <div align="center">$J_{in} = \frac{hC(v)}{\tau_{in}}$ </div>
<br></br>
></br><div align="justify">avec $C(v) = v^2(1-v)$ qui illustre la dépendance du potentiel transmembranaire, et $\tau_{in}$ temps caractéristique de l'application de ce courant.</div>
><br/><div align="justify">Le courant sortant $J_{out}$ est, quant à lui, un combinaison de tous les courants qui diminuent $v$ à l'instar de celui provoqué par le déplacement des ions potassium. Il s'exprime par :</div>
<br></br>
></br> <div align="center"> $J_{out} = -\frac{v}{\tau_{out}}$</div>
<br></br>
></br><div align="justify"> avec ${\tau_{out}}$ durée caractéristique de l'exercice de ce courant.</div>
><br/><div align="justify">Le courant $J_{stim}$ est un courant externe appliqué par l'observateur</div>
><br/><div align="justify">La variable de porte, pour laquelle la valeur 1 correspond au statut ouvert et la valeur 0 au statut fermé, satisfait l'équation différentielle ordinaire suivante :</div>
><br/> <div align="center">$
\frac{dh}{dt} = \left\{
    \begin{array}{ll}
        \frac{1-h}{\tau_{open}} & \mbox{si } v < v_{gate} \\
       \frac{-h}{\tau_{close}} & \mbox{si } v > v_{gate}
    \end{array}
\right.
$</div>

><div align="justify">Il est à noter que ce modèle fait intervenir quatre temps caractéristiques, à savoir $\tau_{in}$, $\tau_{close}$, $\tau_{out}$ et $\tau_{open}$ qui correspondent aux <strong>quatre phases du potentiel d'action</strong> que sont respectivment la <strong>polarisation</strong>, le <strong>plateau</strong>, la <strong>dépolarisation</strong> ainsi que la phase de <strong>retour au potentiel nul</strong>.</div>

#### III.1.2 Hypothèse

><div align="justify">Afin d'obtenir la courbe de restitution dans la section III.2.2 nous travaillerons sous l'hypothèse suivante:<div align="justify">
    ><br/> <div align="center">$\tau_{in} \ll \tau_{out} \ll \tau_{open}, \tau_{close}$</div>

#### III.1.3 Isocline nulle

><div align="justify">L'isocline nulle, caractérisée par l'équation :$\frac{dv}{dt} =0$ permet de mieux apréhender l'évolution couplée de $v$ et $h$.</div>
></br><div align="justify"> En considérant que $J_stim=0$, l'équation d'évolution du potentiel transmembranaire devient alors : </div>
><br/><div align="center">$ 0= \frac{hC(v)}{\tau_{in}} - \frac{v}{\tau_{in}}$</div>
><br/><div align="justify"> On obtient alors, en éliminant les solutions triviales, l'équation de h en fonction de v:</div>
><br/><div align="center">$ h = \frac{\tau_{in}}{\tau_{out}}\frac{1}{v(1-v)}$</div>
><br/><div align="justify">ou encore, en inversant la relation ci-dessus:</div>
><br/><div align="center">$ v = v_{\pm}(h) = \frac{1}{2} \pm \sqrt{\frac{1}{4} - \frac{\tau_{in}}{\tau_{out}}h^{-1}} $</div>
><br/><div align="justify">On remarque alors que $\frac{dv}{dt}$ est positif pour $v_{-}(h) < v < v_{+}(h)$, et au contraire négatif pour $v <v_{-}(h)$ ou $v > v_{+}(h)$.</div>
><br/><div align="justify">On note également qu'au cours de l'isocline nulle, <strong>$h$ atteint son minimum pour $h_{min}= 4\frac{\tau_{in}}{\tau_{out}}$</strong>.</div>

#### III.2.1 Réponse à un stimulus simple

><div align="justify">On conidérera dans cette partie que $J_{stim} =0$ et qu'on partira d'un état initial ($v_{stim}$,1) dans la mesure où la durée d'excitation est faible devant $\tau_{in}$.</div>

```python
tau_close = 150
tau_open = 120
tau_out = 6
tau_in = 0.3


hmin= 4*tau_in/tau_out

v_gate = 13-70
vgate=0.13
v_stim=0.12

v0 = 0.001
vf = 0.999
v1 = np.linspace(v0,vf,100)

t0=0
tf=500
n=2000

t1=np.linspace(t0,tf,n)

plt.plot(v1, tau_in/(tau_out*v1*(1-v1)), 'g--', label="isocline nulle")
plt.xlim([0, 1])
plt.ylim([0,1])

def g(y):
    if y[0]*100-70<v_gate:
        return (1-y[1])/tau_open
    else:
        return -y[1]/y[2]

    
def f_LV2(y,t):
    yp = np.array([(1-y[0])*y[0]*y[0]*y[1]/tau_in - y[0]/tau_out  , g(y),0])
    return yp

y01=np.array([v_stim,1,tau_close])

w1 = itg.odeint(f_LV2,y01, t1)

plt.plot(w1[:,0],w1[:,1],'r',label = "orbite d'un PA")
plt.xlabel("v")
plt.ylabel("h")
plt.legend()
```

><div align="justify">Après avoir tracé l'orbite d'un potentiel d'action ci-dessus, nous traçons désormais le potentiel d'action en question.</div>

```python
plt.plot(t1, w1[:,0], label="potentiel transmembranaire")
plt.xlabel("t (ms)")
plt.ylabel("v")
plt.legend()
```

><div align="justify">Dans les deux graphiques que nous avons obtenus plus haut, se distinguent les quatre phases caractéristiques du potentiel d'action.
><ol>
    <br></br>
    <li>A la suite du stimulus, <strong>$J_{in}$ domine $J_{out}$</strong>, et $v$ augmente rapidement pour atteindre l'isocline nulle $v=v_+(1)$. Cette phase dure de l'ordre de $\tau_{in}$, la variation de h y est négligeable.</li>
    <br></br>
    <li>Une fois la porte fermée, puisque $v > v_{gate}$, $h$ diminue $v$ suit l'isocline nulle et un <strong>équilibre s'établit entre $J_{in}$ et $J_{out}$</strong>. On a alors, pendant une durée de l'ordre de $\tau_{close}$, l'équation $v(t) = v_+(h(t))$.</li>
    <br></br>
    <li>Dès que h atteint $h_{min}$, a solution s'écarte de l'isocline nulle et <strong>$J_{out}$ domine $J_{in}$</strong>, conduisant à une chute du potentiel à 0. Cette phase dure de l'ordre de $\tau_{out}$.</li>
    <br></br>
    <li> $v$ reste très proche de 0 et la <strong>porte se réouvre progressivement</strong> pendant un temps de l'ordre de $\tau_{open}$.</li>
</ol></div>

##### APD

><div align="justify">La forme du potentiel obtenu ou mesuré est souvent mis de côté pour s'intéresser à la <strong>durée du potentiel d'action</strong>, notée par la suite <strong>APD (Action Potentiel Duration)</strong>. Il est défini comme étant la durée au cours de laquelle $v>v_{gate}$. Ceci arrour l'intégralité de la phase 2 ainsi que pour certaines parties des phases 2 et 3. Néanmoins, les temps caractéristiques des phases 1 et 3 étant très faible devant celui de la phase 2 par hypothèse (voir III.1.2.), l' APD est <strong>approximativement égal à la durée de la phase 2</strong>, à savoir de $h=1$ jusqu'à $h=h_{min}$. En intégrant l'équation obtenue en III.1.3., on obtient : </div>
><br/><div align="center">$ APD_{max} = \tau_{close}\ln(\frac{1}{h_{min}})$</div>

>Nous allons chercher à vérifier cette expression.

><div align="justify">Avant de commencer notre étude, notons que h atteint en rélaité son minimum lorsque $v$ redevient inférieur à $v_{gate}$ à la fin de la phase 3. Nous allons donc également vérifier cette expression en remplaçant $h_{min}$ par le réel minimum de $h$.</div>
><div align="justify">La comaparaison de $h_{min}$ et du minimum de $h$ est effectuée ci dessous pour différentes valeurs de $\tau_{close}$.</div>

```python
to1=np.linspace(50,tau_close,n)
HM1=np.array([0.0])
HM2=np.array([0.0])

for tau_cl in to1:
    y0tc=np.array([v_stim,1,tau_cl])
    w1tc = itg.odeint(f_LV2,y0tc, t1)
    
    h_min=w1tc[:,1].min()
    HM1=np.append(HM1,np.array([h_min]))

    HM2=np.append(HM2,np.array([hmin]))
    
    
plt.plot(to1,HM1[1:],'blue', label='$ min(h)$')
plt.plot(to1,HM2[1:],'r', label='$h_{min}$')
plt.legend(prop={'size':12})
plt.xlabel('$t_{close}$')
```

><div align="justify">On observe donc que pour $\tau_{close}$ de l'ordre de la valeur avec laquelle nous travaillons (150ms), le minimum de $h$ et $h_{min}$ sont relativement proches. La différence relative est infèrieure à 10%.</div>

#### III.2.2 Réponse à un stimulus double et courbe de restitution

><div align="justify">Désormais, on excite à nouveau le système revenu au repos après une durée $S$. Cette influence s'observe par l'evolution : $APD_2 = F(DI)$, appelée <strong>courbe de restitution</strong> dont Mitchell-Schaeffer fournit une expression explicite en fonction des paramètres caractéristiques.</div>

><br/><div align="justify">Nous allons observer que la <strong>réponse au deuxième stimulus diffère de la première</strong>, en effet le système ne commence pas à l'équilibre lorsqu'il est stimulé une deuxième fois.</div>

#### V. Le modèle de Fisher-Kolmogorov-Petrovsky-Piskounov (FKPP)

><div align="justify">Désormais, dans le but d'étudier par la suite les équations qui régissent la propagation du potentiel transmembranaire dans le tissu cardiaque, nous allons nous intéresser aux équations de Fisher-Kolmogorov. En effet, on y étudie l'association d'un terme de réaction et de diffusion qui viendra par la suite compléter notre modèle de Mitchell-Schaeffer.</div>

><div align="justify">Ce modèle a été introduit par Fisher en 1937 dans le cadre d'une étude de gêne dominant dans une population. C'est une équation différentielle non linéaire avec un terme de réaction et un terme de diffusion :</div>

>\begin{equation} \label{eq:1}
\frac{\partial{u}}{\partial{t}} = ku(1 - u) + D\frac{\partial^{2}u}{\partial{x^2}}
\end{equation}

>avec $k$ et $D$ des paramétres positive.
Cette équation est le modèle des prototypes qui admet des solutions de front d'onde mobile.

>On opérera à une petite simplification grâce aux changements de variables 
$$
t = kt,  x = x\left(\frac{k}{D}\right)^{1/2}
$$
cela donne finalement,
$$
\begin{array}{c}
\frac{\partial{u}}{\partial{t}} = u(1 - u) + \frac{\partial^{2}u}{\partial{x^2}}
\end{array}
$$

>Une onde stationnaire se déplaçant à la vitesse $c$ est solution particulière de la forme $u(t, x) = U(x - ct).$ En posant $z = x - ct$, $U$ vérifie une équation différentielle ordinaire 
$$
U^{\prime\prime} + cU^{\prime} + U(1 - U) = 0
$$


>On s'intérésse aux ondes stationnaires qui opèrent une transition entre les deux états d'équilibre $U = 1$ et $U = 0$ et vérifiant $0 \le U \le 1.$

><div align="justify">On va étudier cette équation différentielle dans le plan de phase </div>
><br/><div align="center">$$
(U, V) \text{  tels que  } U^{\prime} = V,\text{    }  V^{\prime} = -cV -U(1-U)
$$<div/>
    ><br/><div align="justify">et donc </div>
><br/><div align="center">$$
\frac{dV}{dU} = \frac{-cV - U(1-U)}{V}
$$</div>

>On démontre que pour tout $c \ge 2$ il existe une unique onde stationnaire se déplaçant à la vitesse c. Pour $c < 2$ une telle onde n'existe pas.

```python
""" Importation des bibliothéques """
import numpy as np
import scipy.integrate as itg
import matplotlib.pyplot as plt

zmin=-100
zmax=100

z= np.linspace(zmin,zmax,5000)

def f(y,z):
    u,v,c= y[0], y[1], y[2]
    return np.array([v, -c*v -u*(1-u),0])

#valeurs propres
def lam_p(c):
    return 0.5*(-c+np.sqrt(c**2 -4))

def lam_m(c):
    return 0.5*(-c-np.sqrt(c**2 -4))

c=2.5
nc=10

u0= np.linspace(0.1,1,nc)
v0=np.ones(nc)*(-0.1)
v00=np.ones(nc)*(-0.05)

plt.grid()

for k in range(nc):
    w0=np.array([u0[k],v0[k],c])
    w= itg.odeint(f,w0,z)
    plt.plot(w[:,0],w[:,1])
    
for k in range(nc):
    w0=np.array([u0[k],v00[k],c])
    w= itg.odeint(f,w0,z)
    plt.plot(w[:,0],w[:,1])
    
plt.plot(0,0,'o',color='black')
plt.plot(1,0,'o',color='black')
plt.ylim(-0.12,0.01)
plt.xlim(0,1.1)
plt.xlabel('U')
plt.ylabel("V = U'")
```

>Il y a deux points singuliers pour (U, V ), à savoir (0, 0) et (1, 0).

```python
for k in range(nc):
    w0=np.array([u0[k],v0[k],c])
    w= itg.odeint(f,w0,z)
    plt.plot(z,w[:,0])
    
for k in range(nc):
    w0=np.array([u0[k],v00[k],c])
    w= itg.odeint(f,w0,z)
    plt.plot(z,w[:,0])

plt.grid()
plt.xlabel('z')
plt.ylabel('U(z)')
plt.xlim(-100,-50)
```

>Kolmogorov, Petrovskii et Piscounov (1937) ont démontré que pour toute donnée unitiale $u_0$ vérifiant
><br><br/><div align="center">$
u_0(x) = 1 \text{  } \forall x \le x_1 ,\text{  } u_0(x) = 1 \text{  }\forall x \ge x_2$</div>
><br/><div align="center">$0 \le u_0(x) \le 1
$</div>
><br/><div align="justify">alors la solution $u(x,t)$ converge quant t tend vers l'infini vers une onde stationnaire se déplaçant à la vitesse $c=2$.

>On va étudier l'équation $FKPP$ en discrétisant l'espace et le temps. On va utiliser les formules suivantes 
><br><br/><div align="center">$$
\begin{aligned}
\frac{\partial u}{\partial t}(t, x) & \approx \frac{u(t+k, x)-u(t, x)}{k} \quad \text { progressive } \\
\frac{\partial u}{\partial t}(t, x) & \approx \frac{u(t, x)-u(t-k, x)}{k} \quad \text { retrograde } \\
\frac{\partial^{2} u}{\partial^{2} x}(t, x) & \approx \frac{u(t, x+h)-2 u(t, x)+u(t, x-h)}{h^{2}}
\end{aligned}
$$</div>
><br><div align="justify">On discrétise $[0, T] x [-a, a]$, $\tau = \Delta{t} = \frac{T}{M + 1}$ et $h = \Delta{x} = \frac{2a}{N + 1}$</div>
><br><br/><div align="justify">$$
t_j = j\tau :\text{    } 0 \le j \le M+1, \text{ } x_n = -a + nh : \text{ } 0 \le n \le N+1
$$</div>

>On pose $U_{n}^{j}$ avec $0 \le j \le M+1$ et $0 \le n \le N+1$,  une approximation de $u(t_j, x_n).$

>Suivant qu'on utilise la formule progressive, retrograde ou la moyenne des deux, on aura à resoudre un schema explicite ou implicite.
><br><br/><div align="justify">$$
\begin{array}{c}
\frac{U_{j+1}^{n}-U_{j}^{n}}{\tau}-\frac{U_{j}^{n+1}-2 U_{j}^{n}+U_{j}^{n-1}}{h^{2}}=U_{j}^{n}\left(1-U_{j}^{n}\right) \text{   }(1)\\
\frac{U_{j}^{n}-U_{j-1}^{n}}{k}-\frac{U_{j}^{n+1}-2 U_{j}^{n}+U_{j}^{n-1}}{h^{2}}=U_{j-1}^{n}\left(1-U_{j-1}^{n}\right)   \text{   }(2)
\end{array}
$$</div>


##### V. 1 Schéma Euler explicite

>L'equation (1) nous permettra d'utiliser la méthode Euler explicite. Si $U^j$ désigne le vecteur de taille N et $U_{n}^j$ sa n-ième coordonnée, l'ecriture matricielle donne 
$$
\begin{array}{c}
U^{n+1}=\left(I-\frac{\tau}{h^{2}} A\right) U^{n}+\tau F^{n} \\
\text { avec } F_{j}^{n}=U_{j}^{n}\left(1-U_{j}^{n}\right) \operatorname{si} 2 \leq n \neq N \text { et } F_{1}^{n}=\frac{1}{h^{2}}+U_{1}^{n}\left(1-U_{1}^{n}\right)
\end{array}
$$

>où $I$ représente la matrice identité et $A$ est définie 
$$
A = \left(\begin{array}{ccccc}
2 & -1 & & & \\
-1 & 2 & -1 & & \\
& \ddots & \ddots & \ddots & \\
& & \ddots & \ddots & -1 \\
& & & -1 & 2
\end{array}\right)
$$
><br/><div align="justify">$A$ s'appelle la matrice de Dirichlet.
il faut pour boucler le système (et même l'EDP sur le domaine borné) ajouter des conditions sur le bord. Pour faire simple nous prendrons Dirichlet non homogène $u(t,−a)=1$ et $u(t,a)=0$, en accord avec le théorème.</div>

```python
def schem_expl(y0,T,a,M,N):
    dx, dt = 2*a/(N+1), T/(M+1)
    A=2*np.diag(np.ones(N))-np.diag(np.ones(N-1),-1)-np.diag(np.ones(N-1),1)
    G=np.eye(N)-dt/(dx**2)*A
    Y=np.zeros((M+2,N+2))
    tpt=np.linspace(0,T,M+2)
    xpt=np.linspace(-a,a,N+2)
    Y[0,:]=y0(xpt)
    Y[:,0]=1
    F=np.zeros(N)
    for i in range(M+1):
        F=Y[i,1:-1]*(1-Y[i,1:-1])
        F[0]=F[0]+1/(dx**2)
        Y[i+1,1:-1]=np.dot(G,Y[i,1:-1])+dt*F
    return tpt, xpt, Y

def y1(x):
    return np.ones(np.size(x))*[x<-4]

tpt, xpt, Y = schem_expl(y1,10,10,10000,200)


for i in range(0,tpt.size):
    if i%800==0:
        plt.plot(xpt,Y[i,:])
plt.xlabel('x')
plt.ylabel('U')
plt.show()
```

>Dans cette representation $\Delta{t} = 0.001$ et $\frac{\Delta{t}}{\Delta{x}} = 0.1$ 

```python
tpt, xpt, Y = schem_expl(y1,10,10,2000,200)


for i in range(0,tpt.size):
    if i%800==0:
        plt.plot(xpt,Y[i,:])
plt.xlabel('x')
plt.ylabel('U')
plt.show()
```

>Dans cette representation $\Delta{t} = 0.005$ et $\frac{\Delta{t}}{\Delta{x}} = \frac{1}{2}$. On remarque un probléme de divergence.

>Une analyse plus fine montre que le schéma explicite est stable si et seulement si $\tau = \Delta{t} \le 2$ et $\frac{\Delta{t}}{\Delta{x}^2} \le \frac{1}{2} - \frac{\Delta{t}}{4}.$

```python
import time
def y1(x):
    return np.ones(np.size(x))*[x<-4]

start_time1 = time.time()
tpt, xpt, Y = schem_expl(y1,10,10,10000,500)
time1 = time.time() - start_time1

print(time1)
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2)
ax1.plot(xpt,Y[1,:])
ax1.set_title("j=1")
ax2.plot(xpt,Y[10,:])
ax2.set_title("j=10")
ax3.plot(xpt,Y[20,:])
ax3.set_title("j=20")
ax4.plot(xpt,Y[30,:])
ax4.set_title("j=30")
plt.show()
```

>La méthode d'Euler explicite est trés facile à implenter mais souffre d'un gros defaut de stabilité.

>Nous allons voir une deuxiéme methode, celle du schhéma d'Euler semi implicite. Semi implicite dans le sens où elle sera linéaire parce qu'on évaluera le second membre en $t_{n-1}, x_j$ au lieu de $t_n, x_j$ dans l'équation (2).

##### V. 2 Schéma Euler semi implicite

>Grace à l'equation (2):
$$
\frac{U_{j}^{n}-U_{j}^{n-1}}{k}-\frac{U_{j+1}^{n}-2 U_{j}^{n}+U_{j-1}^{n}}{h^{2}}=U_{j}^{n-1}\left(1-U_{j}^{n-1}\right)
$$
on a 
$$
\begin{array}{c}
\left(I+\frac{\tau}{h^{2}} A\right) U^{(n+1)}=U^{(n)}+\tau F^{(n)} \\
\text { avec } F_{j}^{n}=U_{n, j}\left(1-U_{n, j}\right) \operatorname{si} 2 \leq n \neq N \text { et } F_{1}^{n}=\frac{1}{h^{2}}+U_{1, j}\left(1-U_{1, j}\right)
\end{array}
$$
avec $A$ la même matrice de Dirichlet.

```python
"""importation des bibliotheques"""

import numpy as np
from numpy.linalg import solve
import matplotlib.pyplot as plt
import time


""" definition de la fonction du schema implicite"""
def schem_semi_impl(y0,T,a,M,N):
    """Pas d'espace et de temps"""
    dx, dt = 2*a/(N+1), T/(M+1)
    """Matrice A"""
    A=(2*np.diag(np.ones(N))-np.diag(np.ones(N-1),-1)-np.diag(np.ones(N-1),1))
    """Matrice G = (I+(k/h^2)A)"""
    G=np.eye(N)+dt/(dx**2)*A
    Y=np.zeros((M+2,N+2))
    tpt=np.linspace(0,T,M+2)
    xpt=np.linspace(-a,a,N+2)
    Y[0,:]=y0(xpt)
    Y[:,0]=1
    F=np.zeros(N)
    for i in range(M+1):
        F=Y[i,1:-1]*(1-Y[i,1:-1])
        F[0]=F[0]+1/(dx**2)
        """Resolution du systeme G*U(j+1) = U(j) + tau*F"""
        Y[i+1,1:-1]=solve(G, Y[i,1:-1]+dt*F)
    return tpt, xpt, Y

""" Initialisation """
def y1(x):
    return np.ones(np.size(x))*[x<-4]

tpt, xpt, Y = schem_semi_impl(y1,10,10,1000,800)

for i in range(0,tpt.size):
    if i%80==0:
        plt.plot(xpt,Y[i,:])
plt.show()
```

### IV.4. Comparaison entre la vitesse par la méthode numérique et la vitesse Analytique

#### IV.4.1. Methode

>tel que démmontré analytiquement plus haut, la vitesse de déplacement de l'onde de FKPP est explicitement liée au coéfficients de diffusion et de réaction de l'équation par la formule:
>
>- $\quad \quad \quad \quad V_{min} = 2* \sqrt{k*D}$
>
>Numériquement, étant donné un point $x_1$ situé à une distance a1 du point de discontinuité de $U_0$ (origine de propagation), on récupère l'instant $t_1$ auquel $x_1$ subit une pertubation (valeur de $u(t_1, a_1)>\epsilon$) et on en déduit la vitesse de l'onde par :
>
>-  $\quad \quad \quad \quad vit = \dfrac{a_1}{t_1}$
> 
>en faisant varier la valeur de k, on observe de sur un graphe :
>- en abcisse la vitesse obtenu par la méthode numérique 
>- en ordonnée la vitesse Analytique
>
> dans le cas d'une égalité, la courbe en bleu coinciderait avec la première bissectrice (courbe orange)

```python
import numpy as np
import matplotlib.pyplot as plt
k=10
D=1
t_ar=0

def schem_expl_v(y0,T,a,M,N , k ):
    dx, dt = 2*a/(N+1), T/(M+1)
    #construction de la matrice du système
    A=2*np.diag(np.ones(N))-np.diag(np.ones(N-1),-1)-np.diag(np.ones(N-1),1)
    G=np.eye(N)-(D*dt)/(dx**2)*A
    
    #Initialisation du vecteur de potentielle pour tout temps
    Y=np.zeros((M+2,N+2))
    tpt=np.linspace(0,T,M+2)
    xpt=np.linspace(-a,a,N+2)
    
    Y[0,:]=y0(xpt)
    Y[:,0]=1
    
    F=np.zeros(N)
    #calcul en tout temps de la valeur Y en une position i
    for i in range(M+1):
        v = 0
        t_ar=0
        F=k*Y[i,1:-1]*(1-Y[i,1:-1])
        F[0]=F[0]+D/(dx**2)
        Y[i+1,1:-1]=np.dot(G,Y[i,1:-1])+dt*F 
        
        #détermination de la vitesse de propagation
        if Y[i+1, N-1] > 0.0115:
            v = (14*M)/((i+1)*10)
            t_ar = i-1
            #print (v)
            break
            
    return tpt, xpt, Y , t_ar , v

def y1(x):
    return np.ones(np.size(x))*[x<-4]

tpt, xpt, Y, t_ar , v   = schem_expl_v(y1,10,10,10000,200, k)
```

### V.1 Utilisation d'Odeint

#### V.1.1 Equations du modèle et étude de $v$ et $h$

><br/><div align="justify">Le potentiel transmembranaire suit alors l'équation suivante :</div>
><br/><div align="center">$ \frac{\partial v}{\partial t} =\kappa \frac{\partial^2 v}{\partial x^2} + \frac{h}{\tau_{in}}v^2(1-v) - \frac{v}{\tau_{out}}$</div>
><br/><div align="justify">Quant à la variable de porte, elle vérifie toujours :</div>
><br/><div align="center">$
\frac{dh}{dt} = \left\{
    \begin{array}{ll}
        \frac{1-h}{\tau_{open}} & \mbox{si } v < v_{gate} \\
       \frac{-h}{\tau_{close}} & \mbox{si } v > v_{gate}
    \end{array}
\right.
$</div>

><div align="justify">A l'instar de ce que nous avons fait dans la section précédente, nous allons discrétiser l'espace pour linéariser le laplacien. Nous avons alors un vecteur, dont la première moitié représente $v$ et la deuxième $h$, que nous allons soumettre à la méthode odeint.</div>

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as itg
```

```python
tau_close = 150
tau_open = 120
tau_out = 6
tau_in = 0.3
v_gate = 13-70
#tous les temps sont exprimés en ms

tstim=10
vstim=0.5
inv_omega=500

L=10 #cm
D=0.001 #cm2*ms-1, coefficient de diffusion appelé kappa dans l'équation

T=2000
N=40000

M=60
hp=L/(M+1)
```

><div align="justify">Nous commençons par définir les différentes fonctions qui interviennent dans l'équation de propagation. On consière ici une stimulation périodique en stimulant à un fréquence $w$.</div>

```python
def freac(v,h):
    u=np.zeros(v.size)
    u[:]=h[:]*v[:]*v[:]*(1-v[:])/tau_in - v[:]/tau_out
    return u

def flapl1(v):
    n=v.size
    w=np.zeros(n)
    for i in range(1,n-1):
        w[i]=(v[i+1]+v[i-1]-2*v[i])/(hp**2)
    w[0]=(v[1]-2*v[0])/(hp**2)
    w[n-1]=(v[n-2]-2*v[n-1])/(hp**2)
    
    return w

def Jstim(v,t):
    w=np.zeros(v.size)
    if t%inv_omega<tstim:# or(S<t<S+tstim):
        for i in range((M+2)//(2*L)):
            w[i]=vstim
    return w
    

def g1(v,h):
    return (1-h[:])*[v[:]*100-70<v_gate]/tau_open-(h[:])*[v[:]*100-70>v_gate]/tau_close



def f(y,t):
    v,h=y[:M+2],y[M+2:]
    return np.append(D*flapl1(v) + freac(v,h)+Jstim(v,t), g1(v,h))
```

```python
xpt=np.linspace(-L,L,M+2)
t=np.linspace(0,T,N+2)

v0= np.zeros(np.size(xpt))
h0=np.ones(np.size(xpt))

y0=np.append(v0,h0)

w=itg.odeint(f,y0,t)
```

### V.2. Modele de mitchell et shaeffer : Euler Explicite

#### V.2.1 Discretisation et implémentation
> le modèle de mitchel et shaefer a pour principale particularité qu'il comporte deux variables lié au temps et à l'espace :
>- le potentiel $v(t, x)$
>- le variable de déclenchement $h(t,x)$
> l'équation est donc défini par:
>
>
>$ \qquad\qquad\qquad\qquad\qquad\qquad\qquad\qquad\quad \dfrac{\partial v}{\partial t} = k \dfrac{\partial^2 v}{\partial x^2} + \dfrac{h}{\tau_{in}} v^2 (1-v) - \dfrac{v}{\tau_{out}}$
>
>\begin{equation*} \frac{\partial h}{\partial t} =
\left\{
\begin{aligned}
& \frac{1-h}{\tau_{open}} \quad & si v\leq v_{gate} \\
&\frac{-h}{\tau_{close}},\quad & si v > v_{gate}.
\end{aligned}
\right.\end{equation*}
>
>comme pour le cas FKPP, On pose:
>-$V_{j,n}$ avec $0≤j≤M+1$ et $0≤n≤N+1$ sensé être une approximation de $v(t_j,x_n)$
>- $H_{j,n}$ avec $0≤j≤M+1$ et $0≤n≤N+1$ sensé être une approximation de $h(t_j,x_n)$
>
>pour le cas explicite:
>
>$\dfrac{V_{j+1,n}-V_{j,n}}{dt} -k\dfrac{V_{j,n+1}-2V_{j,n}+V_{j,n-1}}{{dx}^{2}}=\dfrac{H_{j,n}}{\tau_{in}} V^2_{j,n}(1-V_{j,n})- \dfrac{V_{j,n}}{\tau_{out}}$
>
>\begin{equation*} \frac{H_{j+1,n}-H_{j,n}}{dt} =
\left\{
\begin{aligned}
& \frac{1-H_{j,n}}{\tau_{open}} \quad & si V_{j,n}\leq v_crit \\
&\frac{-H_{j,n}}{\tau_{close}},\quad & si V_{j,n}> v_crit.
\end{aligned}
\right.\end{equation*}
>
> 
>de mème que pour FKPP on pose $A$ la matrice du système 
>
>
>$\begin{split}\begin{pmatrix}
2  & -1    &\;     &\;      &\;    \cr
-1  &2    &-1    &\;      &\;    \cr
\:   &\ddots &\ddots &\ddots  &\;    \cr
\: & \:   &\ddots &\ddots & -1    \cr
\;   &\; & \:     &-1     &2    \cr
\end{pmatrix}\end{split}$
>
>
>l'équation sous sa forme matricielle est donc: 
>
> $U^{(j+1)}=(I-\frac{k}{{dx}^2}A)U^{(j)}+ dt F^{(j)}$

> avec $F^{j}_n= \dfrac{H_{j,n}}{\tau_{in}} V^2_{j,n}(1-V_{j,n})- \dfrac{V_{j,n}}{\tau_{out}}$ si $0\leq n\neq N+1$ $

```python
from numpy.linalg import solve

#paramètre du système
D=0.001
tau_close = 150
tau_open = 120
tau_out = 6
tau_in = 0.3
t2 = 5.99
v_gate = 13-70


def schem_expl(y0,T,a,M,N, Jstim):
    dx, dt = 2*a/(N+1), T/(M+1)
    
    #construction de la matrice du système
    A=2*np.diag(np.ones(N+2))-np.diag(np.ones(N+1),-1)-np.diag(np.ones(N+1),1)
    G=np.eye(N+2)-(D*dt)/(dx**2)*A
    
    #initialisation des potentiels et variable de porte
    V=np.zeros((M+2,N+2))
    H=np.zeros((M+2,N+2))
    
    tpt=np.linspace(0,T,M+2)
    xpt=np.linspace(-a,a,N+2)

    V[0,0:10]= 0.12
    H[0,:]=1
  
    F=np.zeros(N+2)
    #détermination en tout temps des variables de potentiel et de porte selon le méthode de discrétisation Euler Explicite  
    for i in range(M+1): 
        F[0:10]=H[i, 0:10]*V[i,0:10]*V[i, 0:10]*(1-V[i,0:10])/tau_in - V[i, 0:10]/tau_out+Jstim[i, 0:10]
        F[10:]=H[i, 10:]*V[i,10:]*V[i, 10:]*(1-V[i,10:])/tau_in - V[i, 10:]/tau_out+Jstim[i, 10:]
        H[i+1, :]= dt *(1-H[i, :])*[V[i, :]*100-70<v_gate]/tau_open-dt*(H[i, :])*[V[i,:]*100-70>v_gate]/tau_close+(H[i, :])
        V[i+1,:]=np.dot(G,V[i,:])+dt*F 
    return tpt, xpt, V , H

```

#### V.2.3. Vitesse de propagation :
>Notre méthode de détermination de la vitesse ayant déja montré son éfficacité dans le cas FKPP, nous la reutiliseront pour la détermination de la vitesse de propagation. pour rappel:
>
>$\dfrac{\partial v}{\partial t} = k \dfrac{\partial^2 v}{\partial x^2} + \dfrac{h}{\tau_{in}} v^2 (1-v) - \dfrac{v}{\tau_{out}}$
>
>comme nous l'avons expliqué c'est l'ajout du laplacien qui a provoqué la propagation dans le domaine; mais ce laplacien est précédé d'un coéffiecient k, nous déterminerons donc la valeur de la vitesse de propagation pour diférentes valeur de k.

>En abcisse on a les valeur de k prise entre [0,1] et en ordonné les valeurs de la vitesse

```python
from numpy.linalg import solve
# paramètre
D=0.001
tau_close = 150
tau_open = 120
tau_out = 6
tau_in = 0.3
t2 = 5.99
v_gate = 13-70
t_ar=0

def schem_expl1(y0,T,a,M,N, Jstim, D):
    dx, dt = 2*a/(N+1), T/(M+1)
    
    A=2*np.diag(np.ones(N+2))-np.diag(np.ones(N+1),-1)-np.diag(np.ones(N+1),1)
    G=np.eye(N+2)-(D*dt)/(dx**2)*A
    
    V=np.zeros((M+2,N+2))
    H=np.zeros((M+2,N+2))
    
    tpt=np.linspace(0,T,M+2)
    xpt=np.linspace(-a,a,N+2)

    V[0,0:10]= 0.12
    H[0,:]=1
  
    F=np.zeros(N+2)
    for i in range(M+1): 
        v = 0
        t_ar=0
        F[0:10]=H[i, 0:10]*V[i,0:10]*V[i, 0:10]*(1-V[i,0:10])/tau_in - V[i, 0:10]/tau_out+Jstim[i, 0:10]
        F[10:]=H[i, 10:]*V[i,10:]*V[i, 10:]*(1-V[i,10:])/tau_in - V[i, 10:]/tau_out+Jstim[i, 10:]
        H[i+1, :]= dt *(1-H[i, :])*[V[i, :]*100-70<v_gate]/tau_open-dt*(H[i, :])*[V[i,:]*100-70>v_gate]/tau_close+(H[i, :])
        V[i+1,:]=np.dot(G,V[i,:])+dt*F 
        
        #calcul de la vitesse de propagation dans le segment pour le modèle de Mitchell et Schaeffer
        if V[i+1, N-10] > 0.0115:
            v = (60*M*2*a)/((i+1)*T*N)
            t_ar = i-1
            break
            
    return tpt, xpt, V , t_ar , v

vinit=0.12
a=20
T=1000
M=30000
N=70
Jstim=np.zeros((M+2,N+2))
vit = np.zeros(100)
D1 = np.linspace(0, 1, 100)
y= np.zeros(100)

from math import sqrt
#détermination de la vitesse en fonction de k
for k in range (100):
    tpt, xpt, Y, t_ar , vit[k]   = schem_expl1(y1,T,a,M,N,Jstim, D1[k])
    y[k]=sqrt(2*D1[k]/tau_in)
plt.plot( y, vit)
#plt.plot( D1, y)
```

>**Observation:** en tracant la valeur de la vitesse en fonction de $\sqrt {\frac{2 k}{\tau_{in}}}$ on obtient une droite. 
>
>la vitesse croit avec la valeur du paramètre k un peu comme $\sqrt { \frac{2 k}{\tau_{in}}}$ ce qui se rapproche assez du cas de la propagation de FKPP. En effet dans le cas de FKPP, pour un D = 1, on avait la relation $v= 2\sqrt k$.

>**interprétation:** la vitesse tend toujours à etre proportionelle à $\sqrt { \frac{2 k}{\tau_{in}}}$ mais dans le cas Mitchell et Schaeffer, c'est un peu plus compliqué par l'intervention de la variable porte et des fonctions de courant. Outre mesure, nous pouvons donc intuité que la valeur de la vitesse de propagation est donné par :
>
>- $v = G()\sqrt { \frac{2 k}{\tau_{in}}} $ 
>
>où G serait une fonction des autres paramètres du système

>En effet, la littérature donne $ v = G(D^*)\sqrt { \frac{2 k}{\tau_{in}}}$ où $D^*$ est la durée d'un potentiel d'action q uand le temps tend vers l'infini.

### V.3.  propagation du potentielle en 2D 

#### V.3.1. **Modélisation et Observation**
> en dehors de l'ajout d'une derivée seconde en espace suivant la deuxième direction, l'équation du système reste inchangé:
>$ \qquad\qquad\qquad\qquad\qquad\qquad\qquad\qquad\quad \dfrac{\partial v}{\partial t} = k (\dfrac{\partial^2 v}{\partial x^2} + \dfrac{\partial^2 v}{\partial y^2})+ \dfrac{h}{\tau_{in}} v^2 (1-v) - \dfrac{v}{\tau_{out}}$
>
>\begin{equation*} \frac{\partial h}{\partial t} =
\left\{
\begin{aligned}
& \frac{1-h}{\tau_{open}} \quad & si v\leq v_{gate} \\
&\frac{-h}{\tau_{close}},\quad & si v > v_{gate}.
\end{aligned}
\right.\end{equation*}
>
>il est donc nécéssaire de prendre en compte cette nouvelle modofication dans l'expréssion dimension dans notre discrétisation:

>$\qquad\qquad\qquad\qquad\quad\dfrac{V^{n+1}_{i,j}-V^n_{i,j}}{dt} -k(\dfrac{V^n_{i+1,j}-2V^n_{i,j}+V^n_{i-1,j}}{{dx}^{2}}+\dfrac{V^n_{i,j+1}-2V^n_{i,j}+V^n_{i,j-1}}{{dy}^{2}})=\dfrac{H^n_{i,j}}{\tau_{in}} (V^2)^n_{i,j}(1-V^n_{i,j})- \dfrac{V^n_{i,j}}{\tau_{out}}$
>
>\begin{equation*} \frac{H^{n+1}_{i,j}-H^n_{i,j}}{dt} =
\left\{
\begin{aligned}
& \frac{1-H^n_{i,j}}{\tau_{open}} \quad & si V^n_{i,j}\leq v_crit \\
&\frac{-H^n_{i,j}}{\tau_{close}},\quad & si V^n_{i,j}> v_crit.
\end{aligned}
\right.\end{equation*}
>

```python
from numpy.linalg import solve
k=2
D=0.01
tau_close = 150
tau_open = 120
tau_in = 0.3
tau_out = np.ones((80+2,80+2))*6
t2 = 5.99
v_gate = 13-70

v0 = 0.001
vf = 0.999
v1 = np.linspace(v0,vf,100)
def schem_expl(y0,T,a,M,N , tau_out):
    dy , dx, dt = 2*a/(N+1), 2*a/(N+1), T/(M+1)
    D=np.ones((N+2,N+2))*0.01
    A=2*np.diag(np.ones(N+2))-np.diag(np.ones(N+1),-1)-np.diag(np.ones(N+1),1)
    G1=np.eye(N+2)-(D*dt)/(dx**2)*A
    G2=-(D*dt)/(dy**2)*A
    
    Y=np.zeros((M+2,N+2,N+2))
    H=np.zeros((M+2,N+2,N+2))
    
    tpt=np.linspace(0,T,M+2)
    xpt=np.linspace(-a,a,N+2)

    Y[0,0:6,0:6]= 0.12
    H[0,:,:]=1
    F=np.zeros((N+2, N+2))
    for i in range(M+1):
        
        F[:,:] = H[i, : ,: ]*Y[i,:,:]*Y[i, :,:]*(1-Y[i,:,:])/tau_in - Y[i, :,:]/tau_out[:,:]
        H[i+1, :, :]= dt *(1 - H[i, :, :]) * [ Y[i, :, :] * 100-70 < v_gate ]/tau_open - dt*(H[i, :, :])*[Y[i,: , :]*100-70>v_gate]/tau_close+ (H[i, :, :])
        Y[i+1,:, :]=np.dot(G1,Y[i,:, :])+np.dot(G2,Y[i,:, :].T).T+dt*F[:,:]
        
    return tpt, xpt, Y , H 

def y1(x):
    return np.ones(np.size(x))*0.12*[x<-7]
tpt, xpt, Y , H = schem_expl(y1,2000,40,20000,80 , tau_out)
```
